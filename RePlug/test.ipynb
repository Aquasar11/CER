```python
from transformers import AutoModel
from transformers import AutoTokenizer

model = AutoModel.from_pretrained("facebook/contriever") 
tokenizer = AutoTokenizer.from_pretrained("facebook/contriever") #Load the associated tokenizer:
```

```python

sentences = [
    "Where was Marie Curie born?",
    "Maria Sklodowska, later known as Marie Curie, was born on November 7, 1867.",
    "Born in Paris on 15 May 1859, Pierre Curie was the son of Eugène Curie, a doctor of French Catholic origin from Alsace."
]

inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors="pt")
embeddings = model(**inputs)
```

```python
def exact_match_score(predictions, references):
    """
    Calculates the Exact Match (EM) score between predictions and references.

    Args:
        predictions (list of str): List of predicted answers.
        references (list of str): List of reference (correct) answers.

    Returns:
        float: Exact Match score as a percentage.
    """
    if len(predictions) != len(references):
        raise ValueError("The number of predictions and references must be the same.")

    exact_matches = sum(pred == ref for pred, ref in zip(predictions, references))
    total = len(references)
    em_score = (exact_matches / total) * 100
    return em_score

# Example usage:
predictions = ["The capital of France is Paris.", "The largest planet is Jupiter."]
references = ["The capital of France is Paris.", "The largest planet is Jupiter."]
score = exact_match_score(predictions, references)
print(f"Exact Match Score: {score:.2f}%")
```
